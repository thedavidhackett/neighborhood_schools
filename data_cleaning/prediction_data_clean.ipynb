```python
#Get only charter schools and neighborhood schools.
stl_schools_filtered = stl_schools[(stl_schools["MAGNET"] == False) & \
                                   (stl_schools["ALTERNATIVE"] == False)]

#Get only needed columns
stl_schools_filtered = stl_schools_filtered[["FULL_CODE", "SCHOOL_NAME", \
                                             "CHARTER", "YEAR", "BGRADE", \
                                             "EGRADE","lat", "lng", "ENROLLMENT_GRADES_K_12"]]
#Get only needed years
stl_schools_filtered = stl_schools_filtered[stl_schools_filtered["YEAR"].isin([2000, 2010])]
```

```python
#Checking for missing data
needed_columns = ['FULL_CODE','SCHOOL_NAME', 'CHARTER', 'BGRADE', 'EGRADE', 'lat', 'lng', \
           "ENROLLMENT_GRADES_K_12", "ENROLLMENT_GRADES_K_12_10"]

stl_schools_00_10[needed_columns].isnull().any()
```

```python
"""Need to fill missing values from schools that weren't in the 2000 data 
and weren't in the 2010 data.
"""
def fill_columns(x):
    columns = ['SCHOOL_NAME', 'CHARTER', 'BGRADE', 'EGRADE', 'lat', 'lng']
    for column in columns:
        if pd.isna(x[column]):
            x[column] = x[column+"_10"]
    if pd.isna(x["ENROLLMENT_GRADES_K_12"]):
        x["ENROLLMENT_GRADES_K_12"] = 0
    if pd.isna(x["ENROLLMENT_GRADES_K_12_10"]):
        x["ENROLLMENT_GRADES_K_12_10"] = 0
        
    return x
```

```python
stl_schools_00_10 = stl_schools_00_10.apply(fill_columns, axis=1)
```

```python
#create flags for elem, middle, high

def create_school_types(x):
    elem = False
    middle =  False
    high = False
    bgrade = x["BGRADE"].strip().lower()
    egrade = x["EGRADE"].strip().lower()
    
    if bgrade == "k" or bgrade == "pk":
        bgrade = 0
    if egrade == "k" or egrade == "pk":
        egrade = 0
        
    bgrade = int(bgrade)
    egrade = int(egrade)
    if bgrade <= 5:
        elem = True
    if bgrade >= 6 and bgrade <= 8 or (elem and egrade >= 6):
        middle = True
    if bgrade >= 9 or egrade == 12:
        high = True
        
    return pd.Series([int(elem), int(middle), int(high)])
```

```python
#Gonna try imputing missing math and english scores based on school type and black pct.

stl_schools_filtered = stl_schools_covariates[stl_schools_covariates["MATH"].notnull()]
eng = stl_schools_filtered["ENG"].to_numpy()
math = stl_schools_filtered["MATH"].to_numpy()
blck = stl_schools_filtered["BLACK_PCT"].to_numpy()
elem = stl_schools_filtered["ELEM"].to_numpy()
middle = stl_schools_filtered["MIDDLE"].to_numpy()
high = stl_schools_filtered["HIGH"].to_numpy()

A = []

for i, value in np.ndenumerate(eng):
    A.append([1, blck[i], elem[i], middle[i], high[i]])
    
beta_eng, _, _, _ = np.linalg.lstsq(A, eng, rcond=None)
beta_math, _, _, _ = np.linalg.lstsq(A, math, rcond=None)

print(beta_eng)
print(beta_math)
```

```python
from math import isnan

def fill_scores(x):
    if isnan(x["MATH"]):
        x["MATH"] = beta_math[0] + beta_math[1] * x["BLACK_PCT"] + x["ELEM"] * beta_math[2] + x["MIDDLE"] * beta_math[3] + x["HIGH"] * beta_math[4]
    if isnan(x["ENG"]):
        x["ENG"] = beta_eng[0] + beta_eng[1] * x["BLACK_PCT"] + x["ELEM"] * beta_eng[2] + x["MIDDLE"] * beta_eng[3] + x["HIGH"] * beta_eng[4]
    return x
```

```python
stl_schools_imputed = stl_schools_covariates.apply(fill_scores, axis=1)
```

```python
stl_schools_imputed["MATH"].isnull().any()
```

```python
# Now to calcuate distances

from math import asin, sqrt, sin, pi, cos

stl_schools_imputed["LATITUDE"] = stl_schools_imputed["lat"] * (pi/180)
stl_schools_imputed["LONGITUDE"] = stl_schools_imputed["lng"] * (pi/180)
stl_schools_imputed.head()
```

```python
# seperate out charters and neighborhood schools

charters = stl_schools_imputed[stl_schools_imputed["CHARTER"] == True]
neighborhood_schools = stl_schools_imputed[stl_schools_imputed["CHARTER"] == False]

def find_distance(lat1, long1, lat2, long2):
    return 2 * 3961 * asin(sqrt(sin((lat2 - lat1)/2)**2 + cos(lat1) * cos(lat2) * sin((long2 - long1)/2)**2))
    
def find_charters(x, distance):
    lat = x["LATITUDE"]
    long = x["LONGITUDE"]
    increase = 0
    for _, charter in charters.iterrows():
        if x["ELEM"] == charter["ELEM"] or x["MIDDLE"] == charter["MIDDLE"] or x["HIGH"] == charter["HIGH"]:
            if find_distance(lat, long, charter["LATITUDE"], charter["LONGITUDE"]) < distance:
                increase += charter["ENROLLMENT_GRADES_K_12_10"]
    return increase
```

```python
neighborhood_schools["CHARTER_CHANGE_1"] = neighborhood_schools.apply(lambda x : \
                                                                      find_charters(x, 1),\
                                                                      axis=1)
```

```python
neighborhood_schools["CHARTER_CHANGE_3"] = neighborhood_schools.apply(lambda x : \
                                                                      find_charters(x, 3),\
                                                                      axis=1)
```
